```python
parsed_table = []
for table in tables:
    for row in table.select("tr")[1:]:
        buffer_row = []
        for column in row.select("td"):
            buffer_row.append(column.get_text())
        # for column in row.select("td")[2:]:
        #     buffer_row.append(int(column.get_text()))
        # parsed_table.append(tuple(buffer_row))
        parsed_table.append(buffer_row)

parsed_table
```
